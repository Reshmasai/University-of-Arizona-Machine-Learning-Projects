```python
# 2b) Provide the following statistics for query string ‚Äúavatar movie‚Äù of top 50 videos
# sorted by relevance in the US region (regionCode)

# search request for top 50 videos sorted by relevance in US region
search_request = youtube_service.search().list(
    q= query_string,
    type= 'video',
    part= 'id,snippet',
    maxResults= 50,
    order = 'relevance',
    regionCode= 'US'
)

# executing the request
response = search_request.execute()

```

```python
from pprint import pprint

#list to store necessary details from response
data = []

# iterating over each result and extracting ID and title
for result in response['items']:
  video_id = result['id']['videoId']
  title = result['snippet']['title']

  # extracting number of (views, likes and comments) using video_id from statistics part
  statistics = youtube_service.videos().list(
       part= 'statistics',
       id= video_id
  )
  statistics_response = statistics.execute()

  #check if necesseray keys are present in the response and assign to variables accordingly
  video_stats = statistics_response['items'][0].get('statistics', {})
  view_count = video_stats.get('viewCount',0)
  like_count = video_stats.get('likeCount',0)
  comment_count = video_stats.get('commentCount',0)

  # appending obtained data to data list
  data.append([video_id, title, view_count, like_count, comment_count])
pprint(data)
```

```python
# 3.c Write a program to export the output of question 3.b in JSON file format and submit the file as part of the assignment

import json

video_ids = list(top_comments['Video ID'])
titles = list(top_comments['Title'])
comment_details = []

for i in range(len(video_ids)):
  video_id = video_ids[i]
  title = titles[i]

  # executing request to retrieve comments
  comments_request = youtube_service.commentThreads().list(
      part = "snippet",
      videoId = video_id,
      maxResults = 10
  )
  comments_response = comments_request.execute()

  # appending necessary data to comment_details list
  comment_details.append({
      'Video ID' : video_id,
      'Title': title,
      'Comments' : comments_response
  })

# giving filename
filename = 'comment_details.json'

# writing comments response to json file
with open(filename,'w') as json_file:
  json.dump(comment_details, json_file, indent=2)

# export message
print(f'comment details exported to {filename}')
```
